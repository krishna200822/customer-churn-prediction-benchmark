```python
X = df.drop(columns=["CustomerID","Churn"])

y = df["Churn"]
```

```python
X.head()
```

```python
X["Spend_Per_Tenure"] = X["Total Spend"] / (X["Tenure"] + 1)
```

```python
X["Support_to_Usage"] = X["Support Calls"] / (X["Usage Frequency"] + 1)
```

```python
X = pd.get_dummies(
    X,
    columns=["Gender","Subscription Type","Contract Length"],
    drop_first=True
)
```

```python
X.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
import numpy as np

# Create model
lr = LogisticRegression()

# 5-fold cross validation
scores = cross_val_score(
    lr,
    X_scaled,
    y,
    cv=5,
    scoring="accuracy"
)

# Fold-wise accuracy
print("Fold Scores:")
print(scores)

# Mean accuracy
print("\nMean Accuracy:")
print(np.mean(scores))

# Standard deviation
print("\nStandard Deviation:")
print(np.std(scores))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier

from sklearn.model_selection import cross_val_score

import pandas as pd
import numpy as np

# Models

models = {

    "LR": LogisticRegression(),

    "DT": DecisionTreeClassifier(),

    "NB": GaussianNB(),

    "KNN": KNeighborsClassifier()

}

results = []

# Benchmark Loop

for name, model in models.items():

    scores = cross_val_score(

        model,

        X_scaled,

        y,

        cv=5,

        scoring="accuracy"

    )

    mean_acc = np.mean(scores)

    std_acc = np.std(scores)

    results.append(

        [name, mean_acc, std_acc]

    )

    print(f"\n{name}")

    print("Fold Scores:", scores)

    print("Mean Accuracy:", mean_acc)

    print("Std Dev:", std_acc)

    print("-"*40)

# Result Table

result_df = pd.DataFrame(

    results,

    columns=[

        "Model",

        "Mean Accuracy",

        "Std Dev"

    ]

)

result_df = result_df.sort_values(

    by="Mean Accuracy",

    ascending=False

)

print("\nFinal Benchmark Table")

print(result_df)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {

    "max_depth":[5,10,15,None],

    "min_samples_split":[2,5,10],

    "min_samples_leaf":[1,2,4]

}

grid = GridSearchCV(

    DecisionTreeClassifier(),

    param_grid,

    cv=3,

    scoring="accuracy",

    n_jobs=-1

)

grid.fit(X_scaled,y)

print(grid.best_params_)

print(grid.best_score_)
```

```python
from sklearn.model_selection import GridSearchCV

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB

# Models + parameter grids

models = {

    "LR":(

        LogisticRegression(),

        {

        "C":[0.01,0.1,1,10],

        "solver":["lbfgs"]

        }

    ),

    "DT":(

        DecisionTreeClassifier(),

        {

        "max_depth":[5,10,15,None],

        "min_samples_split":[2,5],

        "min_samples_leaf":[1,2]

        }

    ),

    "KNN":(

        KNeighborsClassifier(),

        {

        "n_neighbors":[3,5,7],

        "weights":["uniform","distance"]

        }

    ),

    "NB":(

        GaussianNB(),

        {}

    )

}

results = []

for name,(model,param_grid) in models.items():

    grid = GridSearchCV(

        model,

        param_grid,

        cv=3,

        scoring="accuracy",

        n_jobs=-1

    )

    grid.fit(X_scaled,y)

    results.append([

        name,

        grid.best_score_,

        grid.best_params_

    ])

    print("\n",name)

    print("Best Score:",grid.best_score_)

    print("Best Params:",grid.best_params_)

# Result table

result_df = pd.DataFrame(

    results,

    columns=[

    "Model",

    "Best Accuracy",

    "Best Parameters"

    ]

)

result_df = result_df.sort_values(

    by="Best Accuracy",

    ascending=False

)

print(result_df)
```
